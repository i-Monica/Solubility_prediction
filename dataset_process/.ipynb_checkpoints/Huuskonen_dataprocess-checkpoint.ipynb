```python
from rdkit import rdBase, Chem
from rdkit.Chem import AllChem, Draw, PandasTools, Descriptors
from rdkit.Chem.Draw import IPythonConsole
import numpy as np
import pandas as pd
import rdkit
from rdkit.Chem import MACCSkeys
```

```python
df = pd.read_excel('./data/Huuskonen.xlsx')
df['SMILES']
```

```python
mols = []

for smile in df['SMILES']:
    smile = smile.strip()
    mols.append(Chem.MolFromSmiles(smile))

# SMILESmol
mols = [Chem.MolFromSmiles(smile) for smile in df["SMILES"]]

# i=0
# for mol in mols:
# #     print(mol)
#     a = MACCSkeys.GenMACCSKeys(mol)
#     print(a.ToBitString())
#     print(i)
#     i += 1

# Mol
fingerprints = [MACCSkeys.GenMACCSKeys(mol) for mol in mols]
fingerprints = [fingerprint.ToBitString() for fingerprint in fingerprints]
fingerprints
```

```python
# 

new_fingerprints = []

for fingerprint in fingerprints:
    fingerprint = fingerprint.split(' ')
    for single in fingerprint:
        single = list(single)
    new_fingerprints.append(single) 
        
fingerprint_excel = pd.DataFrame(new_fingerprints)

fingerprint_excel.to_csv('./data/Huuskonen_fingerprints.csv', header=None, index=False)
```

```python
df2 = df.join(fingerprint_excel)
df2.to_csv('./data/Huuskonen_gather.csv', index=False)
```
