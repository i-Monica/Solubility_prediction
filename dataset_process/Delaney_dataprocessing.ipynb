```python
from rdkit import rdBase, Chem
from rdkit.Chem import AllChem, Draw, PandasTools, Descriptors
from rdkit.Chem.Draw import IPythonConsole
import numpy as np
import pandas as pd
import rdkit
from rdkit.Chem import MACCSkeys
```

```python
df = pd.read_csv('./data/Delaney.txt', sep=',', encoding='utf-8')
df.columns = ['ID', 'm_sol', 'p_sol', 'SMILES']
```

```python
df1 = df[['ID', 'm_sol', 'SMILES']]
df1.head()
```

```python
# PandasTools.AddMoleculeColumnToFrame(frame=df1, molCol='Mol', smilesCol='SMILES', includeFingerprints=True)
mols = []

for smile in df1['SMILES']:
    mols.append(Chem.MolFromSmiles(smile))

# 将SMILES转化为mol对象
mols = [Chem.MolFromSmiles(smile) for smile in df1["SMILES"]]
```

```python
# 将Mol转化为分子指纹
fingerprints = [MACCSkeys.GenMACCSKeys(mol) for mol in mols]
fingerprints = [fingerprint.ToBitString() for fingerprint in fingerprints]
fingerprints
```

```python
# 指纹转数组

new_fingerprints = []

for fingerprint in fingerprints:
    fingerprint = fingerprint.split(' ')
    for single in fingerprint:
        single = list(single)
    new_fingerprints.append(single) 
        
fingerprint_excel = pd.DataFrame(new_fingerprints)

fingerprint_excel.to_csv('./data/Delaney_fingerprints.csv', header=None, index=False)
```

```python
df2 = df1.join(fingerprint_excel)
df2.to_csv('./data/Delaney_gather.csv', index=False)
```
